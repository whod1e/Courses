# 5) Посчитайте по каждому клиенту среднюю разницу в днях между его пополнениями, выведите медианное значение по всем клиентам.

```python
incomes_sorted = incomes.sort_values("operation_datetime")
```

```python
incomes_sorted['prev'] = incomes_sorted.groupby("user_id")["operation_datetime"].shift(1)
```

```python
incomes_with_prev = incomes_sorted.loc[~incomes_sorted['prev'].isnull()]
```

```python
incomes_with_prev['dif'] = incomes_with_prev['operation_datetime'] - incomes_with_prev['prev']
```

```python
incomes_with_prev_mean = incomes_with_prev.groupby('user_id')['dif'].mean().reset_index()
```

```python
incomes_with_prev_mean = incomes_with_prev_mean.rename(columns = {'user_id':'user_id', 'dif':'avg_dif'})
```

```python
incomes_with_prev_mean['avg_dif_days'] = incomes_with_prev_mean['avg_dif'].dt.days
```

```python
incomes_with_prev_mean['avg_dif'].median()
```
